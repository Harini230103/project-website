# CATBOOST CLASSIFIER ALGORITHEM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('ATTACK.csv')
df.head()
```

```python
df.columns
```

```python
df['Attack_type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['Attack_type',"SERVICE","TTPE"]
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
df['Attack_type'].unique()
```

```python
df=df.dropna()
```

```python
df.tail()
```

```python
x1 = df.drop(labels='Attack_type', axis=1)
y1 = df.loc[:,'Attack_type']
```

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

ros =RandomOverSampler(random_state=42)
x,y=ros.fit_resample(x1,y1)
print("OUR DATASET COUNT         : ", Counter(y1))
print("OVER SAMPLING DATA COUNT  : ", Counter(y))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=6, stratify=y)
print("NUMBER OF TRAIN DATASET    : ", len(x_train))
print("NUMBER OF TEST DATASET      : ", len(x_test))
print("TOTAL NUMBER OF DATASET    : ", len(x_train)+len(x_test))
```

```python
print("NUMBER OF TRAIN DATASET    : ", len(y_train))
print("NUMBER OF TEST DATASET      : ", len(y_test))
print("TOTAL NUMBER OF DATASET    : ", len(y_train)+len(y_test))
```

```python
from catboost import CatBoostClassifier
```

```python
CAT = CatBoostClassifier(random_state=42)
CAT.fit(x_train,y_train)
```

```python
predicted = CAT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test,predicted)
print("THE ACCURACY SCORE OF CATBOOST CLASSIFIER IS :",a*100)
```

```python
from sklearn.metrics import hamming_loss
hl = hamming_loss(y_test,predicted)
print("THE HAMMING LOSS OF CATBOOST CLASSIFIER IS :",hl*100)
```

```python
from sklearn.metrics import classification_report
C = classification_report(y_test,predicted)
print("THE CLASSIFICATION REPORT SCORE OF CATBOOST CLASSIFIER IS :\n\n",C)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(CAT, x, y, scoring='accuracy')
print('THE CROSS VALIDATION TEST RESULT OF ACCURACY :\n\n\n', accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predicted)
print('THE CONFUSION MATRIX SCORE OF CATBOOST CLASSIFIER:\n\n\n',cm)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
CMD = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CAT.classes_)
CMD.plot()
plt.show()
```
